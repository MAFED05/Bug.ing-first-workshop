Leer un número entero y determinar si es par o impar (Punto 3)

```python
# Punto 3
a : int
a = int (input("Ingrese un número entero"))
if a % 2 == 0 :
    print ((str(a)) + " es par")
elif a % 2 == 1 : 
    print (str(a) + " es impar")
```
